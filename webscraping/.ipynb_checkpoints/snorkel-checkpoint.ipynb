```python
import pandas as pd 

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("E:\genai\Youtube01-Psy.csv")
```

```python
df.head()
```

```python
df_train , df_test = train_test_split(df, test_size = 0.2)
```

```python
Y_test = df_test.CLASS.values
```

```python
ABSTAIN = -1
SPAM =1
HAM =0
```

```python
df_train[['AUTHOR', 'CONTENT']].sample(20, random_state = 2)
```

```python
from snorkel.labeling import labeling_function

@labeling_function()
def check(x): 
    return SPAM if "check" in x['CONTENT'].lower() else ABSTAIN

@labeling_function()
def check_out(x): 
    return SPAM  if "check out" in x['CONTENT'].lower() else ABSTAIN
    
```

```python
from snorkel.labeling import PandasLFApplier
lfs = [check_out, check]

applier = PandasLFApplier(lfs=lfs)
L_train = applier.apply(df=df_train)
L_train
```

```python
coverage_check_out, coverage_check = (L_train != ABSTAIN).mean(axis=0)

print(f"check_out_coverage: {coverage_check_out * 100:.1f}%")
print(f"check coverage: {coverage_check * 100:.1f}%")
```

```python
from snorkel.labeling import LFAnalysis

LFAnalysis(L=L_train, lfs=lfs).lf_summary()
```

```python
df_train.iloc[L_train[:,1]==SPAM].sample(10, random_state = 2)
```

```python
# L_train[:,0], L_train[:,1] # first one is for first column and the second one is for second column 

from snorkel.analysis import get_label_buckets

buckets = get_label_buckets(L_train[:, 0], L_train[:, 1])
df_train.iloc[buckets[(ABSTAIN, SPAM)]].sample(10, random_state=1)
```

```python
import re


@labeling_function()
def regex_check_out(x):
    return SPAM if re.search(r"check.*out", x['CONTENT'], flags=re.I) else ABSTAIN
```

```python
lfs = [check_out, check, regex_check_out]

applier = PandasLFApplier(lfs=lfs)
L_train = applier.apply(df=df_train)
```

```python
LFAnalysis(L=L_train, lfs=lfs).lf_summary()
```

```python
buckets = get_label_buckets(L_train[:, 1], L_train[:, 2])
df_train.iloc[buckets[(SPAM, ABSTAIN)]].sample(10, random_state=1)
```

```python
from snorkel.preprocess import preprocessor
from textblob import TextBlob


@preprocessor(memoize=True)
def textblob_sentiment(x):
    scores = TextBlob(x['CONTENT'])
    x.polarity = scores.sentiment.polarity
    x.subjectivity = scores.sentiment.subjectivity
    return x
```

```python
@labeling_function(pre=[textblob_sentiment])
def textblob_polarity(x):
    return HAM if x.polarity > 0.9 else ABSTAIN
    
```

```python
@labeling_function(pre=[textblob_sentiment])
def textblob_subjectivity(x):
    return HAM if x.subjectivity >= 0.5 else ABSTAIN
```

```python
lfs = [textblob_polarity, textblob_subjectivity]

applier = PandasLFApplier(lfs)
L_train = applier.apply(df_train)
```

```python
lfs = [textblob_polarity, textblob_subjectivity]

applier = PandasLFApplier(lfs)
L_train = applier.apply(df_train)
```

```python
 LFAnalysis(L_train, lfs).lf_summary()
```

```python
from snorkel.labeling import LabelingFunction


def keyword_lookup(x, keywords, label):
    if any(word in x["CONTENT"].lower() for word in keywords):
        return label
    return ABSTAIN


def make_keyword_lf(keywords, label=SPAM):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )



"""Spam comments talk about 'my channel', 'my video', etc."""
keyword_my = make_keyword_lf(keywords=["my"])

"""Spam comments ask users to subscribe to their channels."""
keyword_subscribe = make_keyword_lf(keywords=["subscribe"])

"""Spam comments post links to other channels."""
keyword_link = make_keyword_lf(keywords=["http"])

"""Spam comments make requests rather than commenting."""
keyword_please = make_keyword_lf(keywords=["please", "plz"])

"""Ham comments actually talk about the video's content."""
keyword_song = make_keyword_lf(keywords=["song"], label=HAM)
```

```python
@labeling_function()
def short_comment(x):
    """Ham comments are often short, such as 'cool video!'"""
    return HAM if len(x["CONTENT"].split()) < 5 else ABSTAIN
```

```python
lfs = [
    keyword_my,
    keyword_subscribe,
    keyword_link,
    keyword_please,
    keyword_song,
    regex_check_out,
    short_comment,

    textblob_polarity,
    textblob_subjectivity,
]
```

```python
applier = PandasLFApplier(lfs=lfs)
L_train = applier.apply(df=df_train)
L_test = applier.apply(df=df_test)
```

```python
LFAnalysis(L=L_train, lfs=lfs).lf_summary()
```

```python
from snorkel.labeling.model import MajorityLabelVoter

majority_model = MajorityLabelVoter()
preds_train = majority_model.predict(L=L_train)
```

```python
preds_train
```

```python
from snorkel.labeling.model import LabelModel

label_model = LabelModel(cardinality=2, verbose=True)
label_model.fit(L_train=L_train, n_epochs=500, log_freq=100, seed=123)
```

```python
majority_acc = majority_model.score(L=L_test, Y=Y_test, tie_break_policy="random")[
    "accuracy"
]
print(f"{'Majority Vote Accuracy:':<25} {majority_acc * 100:.1f}%")

label_model_acc = label_model.score(L=L_test, Y=Y_test, tie_break_policy="random")[
    "accuracy"
]
print(f"{'Label Model Accuracy:':<25} {label_model_acc * 100:.1f}%")
```

```python
from snorkel.labeling import filter_unlabeled_dataframe

df_train_filtered, probs_train_filtered = filter_unlabeled_dataframe(
    X=df_train, y=preds_train, L=L_train
)
```

```python
import numpy as np
probs_2d = np.stack([1 - probs_train_filtered, probs_train_filtered], axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(ngram_range=(1, 5))
X_train = vectorizer.fit_transform(df_train_filtered['CONTENT'].tolist())
X_test = vectorizer.transform(df_test['CONTENT'].tolist())
```

```python
from snorkel.utils import probs_to_preds

preds_train_filtered = probs_to_preds(probs=probs_2d)
```

```python
from sklearn.linear_model import LogisticRegression

sklearn_model = LogisticRegression(C=1e3, solver="liblinear")
sklearn_model.fit(X=X_train, y=preds_train_filtered)
```

```python
print(f"Test Accuracy: {sklearn_model.score(X=X_test, y=Y_test) * 100:.1f}%")
```
